```python
import matplotlib.pyplot as plt
```

## Stack plot

```python
x=[1,2,3,4,5]
area=[2,2,5,4,3]
area1=[2,8,4,4,6]
area2=[1,3,5,4,6]
area3=[5,6,5,4,7]
l=["area","area1","area2","area3"]
c=["g","r","y","m"]
plt.stackplot(x,area,area1,area2,area3,labels=l,colors=c,baseline="zero")
plt.legend()
plt.title("Python")
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
# plt.grid()
plt.show()
```
