## Stem plot

```python
import matplotlib.pyplot as plt 
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=0,basefmt="g",label="python",use_line_collection=False)
plt.legend()
plt.show()
```

## Oriention

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=0,basefmt="g",label="python",orientation="horizontal")
plt.legend()
plt.show()
```
