## Fill_Between_Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,color="r")
plt.title("Python")
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
# plt.fill_between(x=[2,4],y1=2,y2=3,color="g")
plt.fill_between(x,y,color="g",where=(x>=2) & (x<=4),alpha=0.3)
plt.show()
```
