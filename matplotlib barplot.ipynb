## Bar plot probable playing 11 runs in odi before 02/09/2023

```python
import matplotlib.pyplot as plt
```

```python
X=[9837,1381,12898,1631,1986,1666,2560,164,204,47,30]
Y=["hittu","Gill","VK","Iyer","BKL","har","Jaddu","Kullu","Shami","Siraj","Bumrah"]
```

```python
c=["y","b","m","g","y","b","m","g","y","b","m"]
plt.bar(Y,X,width=0.6,color=c,align="center",edgecolor="r",linewidth=2,linestyle=":",alpha=0.4,label="runs of batter")
plt.legend()
plt.xlabel("Player Name",fontsize=20)
plt.ylabel("Runs",fontsize=20)
plt.title("Player runs before 02/09/2023",fontsize=20)
plt.show()
```

## multiple bar plot

```python
import numpy as np
```

```python
X=[9837,1381,12898,1631,1986,1666,2560,164,204,47,30]
Y=["hittu","Gill","VK","Iyer","BKL","har","Jaddu","Kullu","Shami","Bumrah","Siraj"]
a=[4600.54,6200.48,5700.32,4600.6,4500.14,3300.32,3200.82,1100.71,800.5,600.71,600.0]
width=0.4
p=np.arange(len(X))
p1=[j+width for j in p]
plt.xlabel("Player Name")
plt.ylabel("Runs scored and avg*100")
plt.title("Runs and average of playing 11")
plt.bar(p,X,color="r",width=0.4,label="Runs_scored")
plt.bar(p1,a,color="y",width=0.4,label="Average_of_a_batter")
plt.xticks(p+width/2,Y,rotation=20)
plt.legend()
plt.show()
```
