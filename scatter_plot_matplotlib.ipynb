## scatter plot

```python
import matplotlib.pyplot as plt
```

```python
day=[1,2,3,4,5,6,7]
no=[3,5,9,2,1,4,5]
color=[10,20,30,40,50,60,70]
s=[100,500,400,500,700,1000,700]
plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=20)
plt.scatter(day,no,c=color,s=s,alpha=1.0,marker="*",edgecolor="g",linewidth=1,cmap="BrBG")
t=plt.colorbar()
t.set_label("Color Bar",fontsize=20)
plt.show()
```

## multiple scatter plot

```python
day=[1,2,3,4,5,6,7]
no=[3,5,9,2,1,4,5]
no1=[5,6,9,4,5,7,1]
s=[100,500,400,500,700,1000,700]
plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=20)
plt.scatter(day,no,color="r",s=s,alpha=1.0,marker="*",edgecolor="g",linewidth=1)
plt.scatter(day,no1,color="g",s=s,alpha=1.0,marker="*",edgecolor="g",linewidth=1)
plt.show()
```
