## Box_plot

```python
import matplotlib.pyplot as plt
import random
```

```python
# x=random.sample(range(10, 70), 5)
x=[63, 28, 53, 69, 10,120]
```

```python
plt.boxplot(x,notch=False,vert=True,widths=0.1,labels=["share market"],patch_artist=False,showmeans=True,whis=2,sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),whiskerprops=dict(color="r"),flierprops=dict(markeredgecolor="r"))
plt.show()
```

## multiple boxplot
